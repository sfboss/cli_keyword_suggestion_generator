### 🧠 03. Core Utilities & Logic
This cell contains the **brain** of the text processing engine. It includes functions for intent classification (Commercial vs. Informational) and the *fuzzy logic* used to clean up massive lists of keyword suggestions.

````python

# @title 03_UTILITIES — Cleaning, scoring, CSV/JSON helpers
STOPWORDS = set("""
a an and are as at be but by can do does for from had has have how i if in into is it its near of on or our over the their to vs was were what when where which who why will with without you your
""".split())

QUESTION_STARTERS = {"who", "what", "when", "where", "why", "how", "which", "can", "does", "do", "is", "are", "should", "will"}
COMMERCIAL_TERMS = {
    "best", "top", "buy", "price", "pricing", "cost", "cheap", "deal", "deals", "coupon", "software", "tool", "tools", "service", "services", "consultant", "consulting", "agency", "company", "companies", "platform", "vendor", "vendors", "alternative", "review", "reviews"
}
COMPARISON_TERMS = {"vs", "versus", "alternative", "alternatives", "compare", "comparison", "competitor", "competitors", "or"}
LOCAL_TERMS = {"near me", "nearby", "local", "city", "pittsburgh", "san diego", "new york", "los angeles", "chicago", "austin", "dallas"}
TECH_TERMS = {"api", "integration", "oauth", "rest", "soap", "webhook", "python", "salesforce", "apex", "lwc", "sfdx", "metadata", "mulesoft"}


def normalize_keyword(s: Any) -> str:
    if s is None:
        return ""
    s = str(s).strip().lower()
    s = urllib.parse.unquote_plus(s)
    s = re.sub(r"[\u2018\u2019]", "'", s)
    s = re.sub(r"[\u201c\u201d]", '"', s)
    s = re.sub(r"[^a-z0-9#\+\.\-/&'\s]", " ", s)
    s = re.sub(r"\s+", " ", s).strip()
    return s


def display_keyword(s: str) -> str:
    s = normalize_keyword(s)
    return s


def keyword_tokens(s: str) -> List[str]:
    return [t for t in re.split(r"\s+", normalize_keyword(s)) if t and t not in STOPWORDS]


def should_drop_keyword(s: str) -> bool:
    s_norm = normalize_keyword(s)
    if len(s_norm) < CONFIG.get("min_keyword_chars", 3):
        return True
    if s_norm in {"none", "null", "undefined", "nan"}:
        return True
    for bad in CONFIG.get("drop_keywords_containing", []):
        if normalize_keyword(bad) and normalize_keyword(bad) in s_norm:
            return True
    return False


def write_json(path: Path, data: Any) -> None:
    path.parent.mkdir(parents=True, exist_ok=True)
    with open(path, "w", encoding="utf-8") as f:
        json.dump(data, f, indent=2, ensure_ascii=False, default=str)


def write_text(path: Path, text: str) -> None:
    path.parent.mkdir(parents=True, exist_ok=True)
    with open(path, "w", encoding="utf-8") as f:
        f.write(text)


def read_json(path: Path, default=None):
    if not path.exists():
        return default
    with open(path, "r", encoding="utf-8") as f:
        return json.load(f)


def cache_key(url: str) -> str:
    return hashlib.sha256(url.encode("utf-8")).hexdigest()[:24]


def safe_get(url: str, *, timeout: Optional[int] = None, cache_name: Optional[str] = None) -> Tuple[int, str, Dict[str, str]]:
    timeout = timeout or CONFIG.get("request_timeout_seconds", 12)
    if CONFIG.get("cache_http_responses") and cache_name:
        cached = CACHE_DIR / f"{cache_name}.json"
        if cached.exists():
            data = read_json(cached, {})
            return data.get("status_code", 0), data.get("text", ""), data.get("headers", {})
    try:
        resp = SESSION.get(url, timeout=timeout)
        status, text, headers = resp.status_code, resp.text, dict(resp.headers)
    except Exception as e:
        status, text, headers = 0, json.dumps({"error": str(e)}), {}
    if CONFIG.get("cache_http_responses") and cache_name:
        write_json(CACHE_DIR / f"{cache_name}.json", {"url": url, "status_code": status, "text": text, "headers": headers})
    time.sleep(float(CONFIG.get("request_delay_seconds", 0.25)))
    return status, text, headers


def classify_intent(keyword: str) -> str:
    k = normalize_keyword(keyword)
    toks = set(keyword_tokens(k))
    if any(k.startswith(q + " ") for q in QUESTION_STARTERS) or " how to " in f" {k} ":
        return "informational_question"
    if any(term in k for term in ["near me", "nearby"]) or any(t in toks for t in LOCAL_TERMS):
        return "local"
    if toks & COMPARISON_TERMS or " vs " in f" {k} ":
        return "comparison"
    if toks & COMMERCIAL_TERMS:
        return "commercial"
    if toks & TECH_TERMS:
        return "technical_informational"
    return "informational"


def recommended_content_type(intent: str, keyword: str) -> str:
    k = normalize_keyword(keyword)
    if intent == "commercial":
        if any(x in k for x in ["best", "top", "review", "reviews"]):
            return "comparison/listicle"
        return "service_or_product_landing_page"
    if intent == "comparison":
        return "comparison_page"
    if intent == "local":
        return "local_landing_page"
    if intent == "informational_question":
        return "faq_or_how_to_article"
    if intent == "technical_informational":
        return "technical_guide/tutorial"
    return "educational_article"


def score_keyword(keyword: str, source_count: int = 1, occurrence_count: int = 1) -> float:
    k = normalize_keyword(keyword)
    toks = set(keyword_tokens(k))
    score = 0.0
    score += min(source_count, 5) * 10
    score += min(occurrence_count, 8) * 3
    score += 12 if toks & COMMERCIAL_TERMS else 0
    score += 10 if toks & COMPARISON_TERMS or " vs " in f" {k} " else 0
    score += 8 if any(k.startswith(q + " ") for q in QUESTION_STARTERS) or " how to " in f" {k} " else 0
    score += 6 if toks & TECH_TERMS else 0
    token_len = len(keyword_tokens(k))
    if 3 <= token_len <= 7:
        score += 8
    elif token_len > 10:
        score -= 5
    return round(score, 2)


def fuzzy_dedupe_keywords(df: pd.DataFrame, threshold: int = 94) -> pd.DataFrame:
    if df.empty:
        return df
    df = df.copy()
    df["normalized"] = df["keyword"].map(normalize_keyword)
    df = df.sort_values(["priority_score", "source_count", "occurrence_count"], ascending=False)
    kept = []
    seen = []
    for _, row in tqdm(df.iterrows(), total=len(df), desc="Fuzzy dedupe"):
        norm = row["normalized"]
        if not norm:
            continue
        duplicate = False
        # cheap blocking by first token for speed, then fuzzy match
        first = keyword_tokens(norm)[:1]
        for existing in seen:
            existing_first = keyword_tokens(existing)[:1]
            if first and existing_first and first[0] != existing_first[0]:
                continue
            if fuzz.token_sort_ratio(norm, existing) >= threshold:
                duplicate = True
                break
        if not duplicate:
            kept.append(row)
            seen.append(norm)
    return pd.DataFrame(kept).reset_index(drop=True)
````

### 📡 04. Data Source Registry
We define the various **search engine suggestion APIs** used for research. By registering multiple providers like Google, Bing, and DuckDuckGo, we gather a *diverse set* of real-world user queries.

````python

# @title 04_PROVIDER_REGISTRY — Free suggestion providers + extension point

def parse_json_list(text: str) -> Any:
    try:
        return json.loads(text)
    except Exception:
        return []


def parse_google_firefox(text: str) -> List[str]:
    data = parse_json_list(text)
    if isinstance(data, list) and len(data) > 1 and isinstance(data[1], list):
        return [str(x) for x in data[1]]
    return []


def parse_bing_osjson(text: str) -> List[str]:
    data = parse_json_list(text)
    if isinstance(data, list) and len(data) > 1 and isinstance(data[1], list):
        return [str(x) for x in data[1]]
    return []


def parse_duckduckgo(text: str) -> List[str]:
    data = parse_json_list(text)
    out = []
    if isinstance(data, list):
        for item in data:
            if isinstance(item, dict):
                val = item.get("phrase") or item.get("text")
                if val:
                    out.append(str(val))
            elif isinstance(item, str):
                out.append(item)
    return out


def parse_wikipedia_opensearch(text: str) -> List[str]:
    data = parse_json_list(text)
    if isinstance(data, list) and len(data) > 1 and isinstance(data[1], list):
        return [str(x) for x in data[1]]
    return []


def parse_yahoo(text: str) -> List[str]:
    data = parse_json_list(text)
    try:
        return [str(x.get("key")) for x in data.get("gossip", {}).get("results", []) if x.get("key")]
    except Exception:
        return []


def parse_ecosia(text: str) -> List[str]:
    data = parse_json_list(text)
    out = []
    if isinstance(data, dict):
        for key in ["suggestions", "results"]:
            vals = data.get(key)
            if isinstance(vals, list):
                for v in vals:
                    if isinstance(v, str):
                        out.append(v)
                    elif isinstance(v, dict):
                        out.append(str(v.get("query") or v.get("text") or v.get("phrase") or ""))
    elif isinstance(data, list):
        out.extend([str(x) for x in data if isinstance(x, str)])
    return [x for x in out if x]

PROVIDERS = {
    "google_web": {
        "url": "https://suggestqueries.google.com/complete/search?client=firefox&hl={language}&q={query}",
        "parser": parse_google_firefox,
        "enabled": True,
    },
    "google_youtube": {
        "url": "https://suggestqueries.google.com/complete/search?client=firefox&ds=yt&hl={language}&q={query}",
        "parser": parse_google_firefox,
        "enabled": True,
    },
    "bing": {
        "url": "https://api.bing.com/osjson.aspx?query={query}",
        "parser": parse_bing_osjson,
        "enabled": True,
    },
    "duckduckgo": {
        "url": "https://duckduckgo.com/ac/?q={query}",
        "parser": parse_duckduckgo,
        "enabled": True,
    },
    "wikipedia": {
        "url": "https://en.wikipedia.org/w/api.php?action=opensearch&search={query}&limit=10&namespace=0&format=json",
        "parser": parse_wikipedia_opensearch,
        "enabled": True,
    },
    "yahoo": {
        "url": "https://search.yahoo.com/sugg/gossip/gossip-us-ura/?command={query}&output=json&nresults=20",
        "parser": parse_yahoo,
        "enabled": True,
    },
    "ecosia": {
        "url": "https://ac.ecosia.org/autocomplete?q={query}",
        "parser": parse_ecosia,
        "enabled": True,
    },
}


def add_provider(name: str, url_template: str, parser_func, enabled: bool = True):
    """Extension point: add a provider without touching the workflow."""
    PROVIDERS[name] = {"url": url_template, "parser": parser_func, "enabled": enabled}


def fetch_provider_suggestions(provider_name: str, query: str) -> List[str]:
    spec = PROVIDERS[provider_name]
    encoded_query = urllib.parse.quote_plus(query)
    url = spec["url"].format(query=encoded_query, language=CONFIG.get("language", "en"), country=CONFIG.get("country", "US"))
    status, text, headers = safe_get(url, cache_name=f"provider_{provider_name}_{cache_key(url)}")
    if status != 200:
        return []
    try:
        suggestions = spec["parser"](text)
    except Exception as e:
        print(f"Parser failed for {provider_name}: {e}")
        suggestions = []
    return [display_keyword(x) for x in suggestions if not should_drop_keyword(x)]
````

### 🚀 05. The Modifier Engine
To expand a single seed keyword into **hundreds of opportunities**, this engine applies 'modifiers' like questions and year-based updates. This generates a *broad manifest* of queries to be researched.

````python

# @title 05_MODIFIER_ENGINE — Build seed query universe
CURRENT_YEAR = datetime.now().year


def generate_modifier_queries(seed: str) -> List[Dict[str, str]]:
    seed = normalize_keyword(seed)
    letters = list("abcdefghijklmnopqrstuvwxyz")
    question_templates = [
        "what is {seed}", "how does {seed} work", "how to {seed}", "why use {seed}",
        "when to use {seed}", "who needs {seed}", "can you {seed}", "is {seed} worth it",
        "does {seed} work", "what are the benefits of {seed}", "what are the risks of {seed}",
    ]
    commercial_templates = [
        "best {seed}", "top {seed}", "{seed} pricing", "{seed} cost", "{seed} services",
        "{seed} consultant", "{seed} consulting", "{seed} company", "{seed} software", "{seed} tools",
        "{seed} review", "{seed} reviews",
    ]
    comparison_templates = [
        "{seed} vs", "{seed} versus", "{seed} alternatives", "{seed} comparison", "{seed} competitors",
    ]
    learning_templates = [
        "{seed} guide", "{seed} tutorial", "{seed} examples", "{seed} best practices",
        "{seed} checklist", "{seed} course", "learn {seed}", "{seed} for beginners",
        "{seed} interview questions", "{seed} documentation",
    ]
    pain_templates = [
        "{seed} problems", "{seed} issues", "{seed} mistakes", "{seed} not working",
        "{seed} error", "{seed} limitations", "{seed} risks", "{seed} help",
    ]
    freshness_templates = [
        f"{{seed}} {CURRENT_YEAR}", "latest {seed}", "new {seed}", "{seed} trends", "future of {seed}",
    ]
    local_templates = [
        "{seed} near me", "local {seed}", "{seed} agency", "{seed} firm",
    ]
    az_suffix = [f"{seed} {letter}" for letter in letters]
    az_prefix = [f"{letter} {seed}" for letter in letters[:8]]  # smaller prefix set; suffix is usually stronger

    groups = {
        "seed": [seed],
        "a_z_suffix": az_suffix,
        "a_z_prefix_light": az_prefix,
        "questions": [t.format(seed=seed) for t in question_templates],
        "commercial": [t.format(seed=seed) for t in commercial_templates],
        "comparison": [t.format(seed=seed) for t in comparison_templates],
        "learning": [t.format(seed=seed) for t in learning_templates],
        "pain": [t.format(seed=seed) for t in pain_templates],
        "freshness": [t.format(seed=seed) for t in freshness_templates],
        "local": [t.format(seed=seed) for t in local_templates],
    }
    rows = []
    for group, queries in groups.items():
        for q in queries:
            qn = normalize_keyword(q)
            if qn and not should_drop_keyword(qn):
                rows.append({"seed_keyword": seed, "query": qn, "modifier_group": group})
    # stable dedupe / cap
    seen = set()
    unique = []
    for r in rows:
        if r["query"] not in seen:
            unique.append(r)
            seen.add(r["query"])
    return unique[: int(CONFIG.get("max_seed_queries_per_seed", 90))]


def build_all_seed_queries(seed_keywords: List[str]) -> pd.DataFrame:
    rows = []
    for seed in seed_keywords:
        rows.extend(generate_modifier_queries(seed))
    return pd.DataFrame(rows)

seed_query_df = build_all_seed_queries(CONFIG["seed_keywords"])
seed_query_df.to_csv(MASTER_DIR / "seed_query_manifest.csv", index=False)
print("Seed queries generated:", len(seed_query_df))
seed_query_df.head(10)
````

### 🚜 06. Keyword Harvesting & Refinement
This stage executes the actual **data collection**. It loops through queries, fetches live suggestions, and performs *multi-pass deduplication* to produce a high-quality master list.

````python

# @title 06_KEYWORD_UNIVERSE — Fetch, cache, dedupe, score

def collect_suggestions_for_queries(seed_query_df: pd.DataFrame, pass_name: str = "pass1") -> pd.DataFrame:
    rows = []
    enabled_providers = [name for name, spec in PROVIDERS.items() if spec.get("enabled")]
    total = len(seed_query_df) * len(enabled_providers)
    pbar = tqdm(total=total, desc=f"Collect suggestions {pass_name}")
    for _, qrow in seed_query_df.iterrows():
        query = qrow["query"]
        seed = qrow.get("seed_keyword", query)
        modifier_group = qrow.get("modifier_group", "unknown")
        for provider in enabled_providers:
            suggestions = fetch_provider_suggestions(provider, query)
            for sug in suggestions:
                rows.append({
                    "pass": pass_name,
                    "seed_keyword": seed,
                    "source_query": query,
                    "modifier_group": modifier_group,
                    "source": provider,
                    "keyword": display_keyword(sug),
                    "normalized": normalize_keyword(sug),
                })
            pbar.update(1)
    pbar.close()
    df = pd.DataFrame(rows)
    if df.empty:
        return pd.DataFrame(columns=["pass", "seed_keyword", "source_query", "modifier_group", "source", "keyword", "normalized"])
    df = df[~df["normalized"].map(should_drop_keyword)].drop_duplicates()
    return df.reset_index(drop=True)


def build_deduped_keyword_table(raw_df: pd.DataFrame) -> pd.DataFrame:
    if raw_df.empty:
        return pd.DataFrame(columns=["keyword", "normalized", "source_count", "sources", "occurrence_count", "seed_keywords", "modifier_groups", "intent", "content_type", "priority_score"])
    grouped = raw_df.groupby("normalized").agg(
        keyword=("keyword", "first"),
        source_count=("source", lambda x: len(set(x))),
        sources=("source", lambda x: ", ".join(sorted(set(x)))),
        occurrence_count=("keyword", "count"),
        seed_keywords=("seed_keyword", lambda x: ", ".join(sorted(set(map(str, x))))),
        source_queries=("source_query", lambda x: " | ".join(list(dict.fromkeys(map(str, x)))[:8])),
        modifier_groups=("modifier_group", lambda x: ", ".join(sorted(set(map(str, x))))),
    ).reset_index()
    grouped["intent"] = grouped["keyword"].map(classify_intent)
    grouped["content_type"] = grouped.apply(lambda r: recommended_content_type(r["intent"], r["keyword"]), axis=1)
    grouped["priority_score"] = grouped.apply(lambda r: score_keyword(r["keyword"], r["source_count"], r["occurrence_count"]), axis=1)
    grouped = grouped.sort_values(["priority_score", "source_count", "occurrence_count", "keyword"], ascending=[False, False, False, True])
    if CONFIG.get("enable_fuzzy_dedupe") and len(grouped) > 1:
        grouped = fuzzy_dedupe_keywords(grouped, int(CONFIG.get("fuzzy_similarity_threshold", 94)))
    return grouped.reset_index(drop=True)


def write_per_seed_outputs(raw_df: pd.DataFrame, deduped_df: pd.DataFrame) -> None:
    for seed in CONFIG["seed_keywords"]:
        seed_norm = normalize_keyword(seed)
        seed_slug = slugify(seed_norm)
        seed_dir = RUN_DIR / seed_slug
        seed_dir.mkdir(parents=True, exist_ok=True)
        seed_raw = raw_df[raw_df["seed_keyword"].map(normalize_keyword) == seed_norm].copy() if not raw_df.empty else raw_df.copy()
        seed_norms = set(seed_raw["normalized"].unique()) if not seed_raw.empty else set()
        seed_deduped = deduped_df[deduped_df["normalized"].isin(seed_norms)].copy() if seed_norms else pd.DataFrame(columns=deduped_df.columns)
        seed_raw.to_csv(seed_dir / "raw_suggestions.csv", index=False)
        seed_deduped.to_csv(seed_dir / "deduped_keywords.csv", index=False)
        write_text(seed_dir / "deduped_keywords.txt", "\n".join(seed_deduped.get("keyword", pd.Series(dtype=str)).astype(str).tolist()))
        source_summary = seed_raw.groupby("source").agg(rows=("keyword", "count"), unique_keywords=("normalized", "nunique")).reset_index() if not seed_raw.empty else pd.DataFrame(columns=["source", "rows", "unique_keywords"])
        source_summary.to_csv(seed_dir / "source_summary.csv", index=False)


def run_keyword_universe() -> Tuple[pd.DataFrame, pd.DataFrame]:
    raw1 = collect_suggestions_for_queries(seed_query_df, "pass1")
    raw_frames = [raw1]
    deduped1 = build_deduped_keyword_table(raw1)

    if CONFIG.get("enable_second_pass") and not deduped1.empty:
        top_terms = deduped1.head(int(CONFIG.get("second_pass_limit", 25)))["keyword"].tolist()
        second_rows = []
        for term in top_terms:
            second_rows.append({"seed_keyword": term, "query": term, "modifier_group": "second_pass_top_keyword"})
            second_rows.append({"seed_keyword": term, "query": f"how to {term}", "modifier_group": "second_pass_question"})
            second_rows.append({"seed_keyword": term, "query": f"best {term}", "modifier_group": "second_pass_commercial"})
        second_df = pd.DataFrame(second_rows).drop_duplicates("query")
        raw2 = collect_suggestions_for_queries(second_df, "pass2")
        raw_frames.append(raw2)

    raw_all = pd.concat(raw_frames, ignore_index=True).drop_duplicates() if raw_frames else pd.DataFrame()
    deduped = build_deduped_keyword_table(raw_all)

    raw_all.to_csv(MASTER_DIR / "raw_suggestions.csv", index=False)
    deduped.to_csv(MASTER_DIR / "deduped_keywords.csv", index=False)
    write_text(MASTER_DIR / "deduped_keywords.txt", "\n".join(deduped.get("keyword", pd.Series(dtype=str)).astype(str).tolist()))

    source_summary = raw_all.groupby("source").agg(rows=("keyword", "count"), unique_keywords=("normalized", "nunique")).reset_index() if not raw_all.empty else pd.DataFrame(columns=["source", "rows", "unique_keywords"])
    source_summary.to_csv(MASTER_DIR / "source_summary.csv", index=False)

    money = deduped[deduped["intent"].isin(["commercial", "comparison", "local"])].copy()
    money.to_csv(MASTER_DIR / "money_keywords.csv", index=False)

    write_per_seed_outputs(raw_all, deduped)
    return raw_all, deduped
````

### 📊 07. AI Clustering & Strategic Roadmap
Using machine learning, this section groups keywords into **topical clusters**. It identifies the primary keyword for each group and builds a *strategic roadmap* prioritized by commercial value.

````python

# @title 07_CLUSTERING_AND_ROADMAP — Cluster keywords, label topics, produce roadmap

def choose_cluster_count(n: int) -> int:
    if not CONFIG.get("cluster_count_auto"):
        return max(2, int(CONFIG.get("cluster_count", 12)))
    if n < 20:
        return max(2, min(4, n))
    hint = max(4, int(math.sqrt(n / max(1, CONFIG.get("min_cluster_size_hint", 8))) * 4))
    return int(max(4, min(30, hint, max(2, n // 5))))


def tfidf_cluster_keywords(deduped_df: pd.DataFrame) -> pd.DataFrame:
    df = deduped_df.copy()
    if df.empty:
        df["cluster_id"] = []
        df["cluster_label"] = []
        return df
    docs = df["keyword"].fillna("").astype(str).tolist()
    if len(docs) < 3:
        df["cluster_id"] = 0
        df["cluster_label"] = "misc"
        return df

    vectorizer = TfidfVectorizer(ngram_range=(1, 3), min_df=1, stop_words="english", max_features=6000)
    X = vectorizer.fit_transform(docs)
    k = choose_cluster_count(len(df))
    k = max(2, min(k, len(df) - 1))
    model = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = model.fit_predict(X)
    df["cluster_id"] = labels

    terms = np.array(vectorizer.get_feature_names_out())
    cluster_labels = {}
    for cid in sorted(set(labels)):
        idx = np.where(labels == cid)[0]
        centroid = np.asarray(X[idx].mean(axis=0)).ravel()
        top_term_idx = centroid.argsort()[::-1][:6]
        label_terms = [terms[i] for i in top_term_idx if terms[i] not in STOPWORDS]
        cluster_labels[cid] = " / ".join(label_terms[:3]) if label_terms else f"cluster_{cid}"
    df["cluster_label"] = df["cluster_id"].map(cluster_labels)
    return df


def embedding_cluster_keywords(deduped_df: pd.DataFrame) -> pd.DataFrame:
    if not CONFIG.get("enable_embeddings"):
        return tfidf_cluster_keywords(deduped_df)
    try:
        from sentence_transformers import SentenceTransformer
        from sklearn.preprocessing import normalize as sk_normalize
        model = SentenceTransformer("all-MiniLM-L6-v2")
        docs = deduped_df["keyword"].fillna("").astype(str).tolist()
        emb = model.encode(docs, show_progress_bar=True)
        emb = sk_normalize(emb)
        k = max(2, min(choose_cluster_count(len(docs)), len(docs) - 1))
        km = KMeans(n_clusters=k, random_state=42, n_init=10)
        labels = km.fit_predict(emb)
        df = deduped_df.copy()
        df["cluster_id"] = labels
        # label clusters via TFIDF within each cluster
        tfidf_labeled = tfidf_cluster_keywords(df[["keyword", "normalized", "source_count", "sources", "occurrence_count", "seed_keywords", "source_queries", "modifier_groups", "intent", "content_type", "priority_score"]].copy())
        label_map = tfidf_labeled.groupby("cluster_id")["cluster_label"].first().to_dict()
        df["cluster_label"] = df["cluster_id"].map(label_map).fillna("cluster")
        return df
    except Exception as e:
        print("Embedding clustering failed; falling back to TF-IDF:", e)
        return tfidf_cluster_keywords(deduped_df)


def build_content_roadmap(clustered_df: pd.DataFrame) -> pd.DataFrame:
    if clustered_df.empty:
        return pd.DataFrame()
    rows = []
    for cid, g in clustered_df.groupby("cluster_id"):
        g = g.sort_values("priority_score", ascending=False)
        primary = g.iloc[0]
        intents = g["intent"].value_counts().to_dict()
        dominant_intent = max(intents, key=intents.get)
        content_type = recommended_content_type(dominant_intent, primary["keyword"])
        supporting = g["keyword"].head(20).tolist()
        questions = [kw for kw in supporting if classify_intent(kw) == "informational_question"][:8]
        commercial_count = int(g["intent"].isin(["commercial", "comparison", "local"]).sum())
        avg_score = round(float(g["priority_score"].mean()), 2)
        max_score = round(float(g["priority_score"].max()), 2)
        priority = round(max_score + min(len(g), 40) * 0.8 + commercial_count * 1.5, 2)
        title = make_title(primary["keyword"], dominant_intent)
        rows.append({
            "cluster_id": cid,
            "cluster_label": primary["cluster_label"],
            "primary_keyword": primary["keyword"],
            "recommended_title": title,
            "url_slug": slugify(primary["keyword"]).replace("_", "-"),
            "dominant_intent": dominant_intent,
            "recommended_content_type": content_type,
            "keyword_count": len(g),
            "commercial_keyword_count": commercial_count,
            "avg_keyword_score": avg_score,
            "max_keyword_score": max_score,
            "roadmap_priority_score": priority,
            "supporting_keywords": " | ".join(supporting),
            "faq_candidates": " | ".join(questions),
            "recommended_schema": recommended_schema_for_content(content_type, questions),
        })
    return pd.DataFrame(rows).sort_values("roadmap_priority_score", ascending=False).reset_index(drop=True)


def make_title(keyword: str, intent: str) -> str:
    k = keyword.strip()
    if intent == "comparison":
        return f"{k.title()}: Comparison, Alternatives, and Decision Guide"
    if intent == "commercial":
        return f"Best {k.title()}: Options, Costs, and How to Choose"
    if intent == "local":
        return f"{k.title()}: Local Options, Costs, and Questions to Ask"
    if intent == "technical_informational":
        return f"{k.title()}: Technical Guide, Examples, and Best Practices"
    if intent == "informational_question":
        return f"{k[0].upper() + k[1:] if k else k}: Clear Answer and Practical Examples"
    return f"{k.title()}: Complete Guide"


def recommended_schema_for_content(content_type: str, questions: List[str]) -> str:
    schemas = ["Article", "BreadcrumbList"]
    if questions:
        schemas.append("FAQPage")
    if "how_to" in content_type or "tutorial" in content_type:
        schemas.append("HowTo")
    if "product" in content_type or "service" in content_type:
        schemas.append("Service")
    if "local" in content_type:
        schemas.append("LocalBusiness")
    return ", ".join(dict.fromkeys(schemas))


def run_clustering_and_roadmap(deduped_df: pd.DataFrame) -> Tuple[pd.DataFrame, pd.DataFrame]:
    clustered = embedding_cluster_keywords(deduped_df)
    clustered.to_csv(MASTER_DIR / "keyword_clusters.csv", index=False)
    roadmap = build_content_roadmap(clustered)
    roadmap.to_csv(MASTER_DIR / "content_roadmap.csv", index=False)
    return clustered, roadmap
````

### 🔍 08. Search Console Integration (Optional)
If a GSC export is provided, this cell analyzes **live ranking data**. It identifies *'striking distance'* opportunities where small content updates could lead to significant traffic gains.

````python

# @title 08_GSC_OPPORTUNITY_MINER — Optional Search Console CSV analysis
GSC_COLUMN_ALIASES = {
    "query": ["query", "queries", "search term", "search_term", "keyword"],
    "page": ["page", "url", "landing page", "landing_page", "pages"],
    "clicks": ["clicks", "click"],
    "impressions": ["impressions", "impression"],
    "ctr": ["ctr", "click through rate", "click_through_rate"],
    "position": ["position", "average position", "avg position", "avg_position", "average_position"],
}


def normalize_columns(df: pd.DataFrame) -> pd.DataFrame:
    col_map = {}
    lower_to_original = {c.strip().lower(): c for c in df.columns}
    for canonical, aliases in GSC_COLUMN_ALIASES.items():
        for alias in aliases:
            if alias in lower_to_original:
                col_map[lower_to_original[alias]] = canonical
                break
    return df.rename(columns=col_map)


def parse_ctr(x) -> float:
    if pd.isna(x):
        return 0.0
    s = str(x).strip().replace("%", "")
    try:
        val = float(s)
        if val > 1:
            val = val / 100.0
        return val
    except Exception:
        return 0.0


def expected_ctr(position: float) -> float:
    # Conservative curve for prioritization, not a claim of universal CTR.
    if position <= 1: return 0.28
    if position <= 2: return 0.16
    if position <= 3: return 0.10
    if position <= 5: return 0.06
    if position <= 8: return 0.035
    if position <= 10: return 0.025
    if position <= 20: return 0.01
    return 0.004


def load_gsc_csv(path: str) -> pd.DataFrame:
    if not path:
        return pd.DataFrame()
    p = Path(path)
    if not p.exists():
        print("GSC CSV not found:", path)
        return pd.DataFrame()
    df = pd.read_csv(p)
    df = normalize_columns(df)
    required = {"query", "page", "clicks", "impressions", "ctr", "position"}
    missing = required - set(df.columns)
    if missing:
        print("GSC CSV missing columns:", missing)
        print("Found columns:", list(df.columns))
        return pd.DataFrame()
    df["query"] = df["query"].map(normalize_keyword)
    df["page"] = df["page"].astype(str)
    df["clicks"] = pd.to_numeric(df["clicks"], errors="coerce").fillna(0)
    df["impressions"] = pd.to_numeric(df["impressions"], errors="coerce").fillna(0)
    df["ctr"] = df["ctr"].map(parse_ctr)
    df["position"] = pd.to_numeric(df["position"], errors="coerce").fillna(99)
    return df


def classify_gsc_opportunity(row) -> str:
    pos = row["position"]
    imps = row["impressions"]
    ctr = row["ctr"]
    exp = expected_ctr(pos)
    if imps >= 500 and 8 <= pos <= 20:
        return "striking_distance_refresh"
    if imps >= 500 and pos <= 10 and ctr < exp * 0.55:
        return "low_ctr_title_meta_rewrite"
    if imps >= 200 and pos > 20:
        return "new_or_expanded_content_needed"
    if imps >= 100 and row["clicks"] == 0:
        return "zero_click_query_review"
    return "monitor"


def gsc_priority_score(row) -> float:
    imps = math.log1p(max(row["impressions"], 0)) * 8
    pos_bonus = max(0, 25 - row["position"]) * 1.5 if row["position"] <= 25 else 0
    ctr_gap = max(0, expected_ctr(row["position"]) - row["ctr"]) * 100
    opp_bonus = {
        "striking_distance_refresh": 25,
        "low_ctr_title_meta_rewrite": 22,
        "new_or_expanded_content_needed": 15,
        "zero_click_query_review": 12,
        "monitor": 3,
    }.get(row.get("opportunity_type"), 0)
    return round(imps + pos_bonus + ctr_gap + opp_bonus, 2)


def analyze_gsc_opportunities(path: str) -> pd.DataFrame:
    df = load_gsc_csv(path)
    if df.empty:
        return pd.DataFrame()
    df["expected_ctr"] = df["position"].map(expected_ctr)
    df["ctr_gap"] = df["expected_ctr"] - df["ctr"]
    df["opportunity_type"] = df.apply(classify_gsc_opportunity, axis=1)
    df["gsc_priority_score"] = df.apply(gsc_priority_score, axis=1)

    # Cannibalization: same query with multiple pages
    cannibal = df.groupby("query")["page"].nunique().reset_index(name="ranking_page_count")
    df = df.merge(cannibal, on="query", how="left")
    df["possible_cannibalization"] = df["ranking_page_count"] > 1
    df["recommended_action"] = df["opportunity_type"].map({
        "striking_distance_refresh": "Refresh page, improve answer block, add missing sections, add internal links.",
        "low_ctr_title_meta_rewrite": "Rewrite title/meta for clearer benefit, specificity, and query match.",
        "new_or_expanded_content_needed": "Create or expand content targeting this query cluster.",
        "zero_click_query_review": "Check SERP intent; consider concise answer/FAQ or skip if zero-click risk is high.",
        "monitor": "Keep monitoring; not a top action yet.",
    })
    df = df.sort_values("gsc_priority_score", ascending=False)
    df.to_csv(MASTER_DIR / "gsc_opportunities.csv", index=False)

    cannibal_df = df[df["possible_cannibalization"]].copy()
    cannibal_df.to_csv(MASTER_DIR / "gsc_cannibalization_candidates.csv", index=False)
    return df
````

### 🕷️ 09. Technical Site Audit (Optional)
This lightweight crawler visits your site's URLs to check for **SEO health**. It flags issues like missing meta descriptions or *thin content*, providing a quick technical overview.

````python

# @title 09_CRAWL_LITE — Optional lightweight technical/content crawl

def normalize_url(url: str, base: str = "") -> str:
    if base:
        url = urllib.parse.urljoin(base, url)
    parts = urllib.parse.urlsplit(url)
    if not parts.scheme:
        return ""
    clean = urllib.parse.urlunsplit((parts.scheme, parts.netloc, parts.path.rstrip("/") or "/", "", ""))
    return clean


def get_sitemap_urls(site_url: str, limit: int) -> List[str]:
    site_url = site_url.rstrip("/")
    candidates = [f"{site_url}/sitemap.xml", f"{site_url}/sitemap_index.xml"]
    urls = []
    for sm in candidates:
        status, text, _ = safe_get(sm, timeout=12, cache_name=f"sitemap_{cache_key(sm)}")
        if status != 200 or not text.strip():
            continue
        try:
            root = ET.fromstring(text.encode("utf-8"))
            locs = [el.text.strip() for el in root.iter() if el.tag.endswith("loc") and el.text]
            for loc in locs:
                if loc.endswith(".xml") and len(urls) < limit:
                    s2, t2, _ = safe_get(loc, timeout=12, cache_name=f"sitemap_child_{cache_key(loc)}")
                    if s2 == 200:
                        try:
                            root2 = ET.fromstring(t2.encode("utf-8"))
                            urls.extend([el.text.strip() for el in root2.iter() if el.tag.endswith("loc") and el.text])
                        except Exception:
                            pass
                else:
                    urls.append(loc)
                if len(urls) >= limit:
                    break
        except Exception:
            continue
        if urls:
            break
    # fallback to homepage only
    if not urls and site_url:
        urls = [site_url]
    seen = []
    for u in urls:
        nu = normalize_url(u)
        if nu and nu not in seen:
            seen.append(nu)
    return seen[:limit]


def crawl_page(url: str, site_netloc: str) -> Dict[str, Any]:
    status, text, headers = safe_get(url, timeout=15, cache_name=f"crawl_{cache_key(url)}")
    row = {"url": url, "status_code": status, "content_type": headers.get("Content-Type", "")}
    if status != 200 or "text/html" not in row["content_type"] and "html" not in text[:200].lower():
        return row
    soup = BeautifulSoup(text, "html.parser")
    title = soup.title.get_text(" ", strip=True) if soup.title else ""
    meta_desc_tag = soup.find("meta", attrs={"name": re.compile("description", re.I)})
    canonical_tag = soup.find("link", rel=lambda x: x and "canonical" in x)
    robots_tag = soup.find("meta", attrs={"name": re.compile("robots", re.I)})
    h1s = [h.get_text(" ", strip=True) for h in soup.find_all("h1")]
    h2s = [h.get_text(" ", strip=True) for h in soup.find_all("h2")]
    text_content = soup.get_text(" ", strip=True)
    words = re.findall(r"\b\w+\b", text_content)
    scripts_jsonld = soup.find_all("script", type="application/ld+json")
    links = []
    for a in soup.find_all("a", href=True):
        nu = normalize_url(a["href"], url)
        if nu:
            links.append(nu)
    internal = [u for u in links if urllib.parse.urlsplit(u).netloc == site_netloc]
    external = [u for u in links if urllib.parse.urlsplit(u).netloc != site_netloc]
    imgs = soup.find_all("img")
    missing_alt = sum(1 for img in imgs if not img.get("alt"))
    row.update({
        "title": title,
        "title_len": len(title),
        "meta_description": meta_desc_tag.get("content", "").strip() if meta_desc_tag else "",
        "meta_description_len": len(meta_desc_tag.get("content", "").strip()) if meta_desc_tag else 0,
        "canonical": canonical_tag.get("href", "").strip() if canonical_tag else "",
        "robots_meta": robots_tag.get("content", "").strip() if robots_tag else "",
        "noindex": "noindex" in (robots_tag.get("content", "").lower() if robots_tag else ""),
        "h1_count": len(h1s),
        "h1": " | ".join(h1s[:3]),
        "h2_count": len(h2s),
        "word_count": len(words),
        "jsonld_count": len(scripts_jsonld),
        "image_count": len(imgs),
        "images_missing_alt": missing_alt,
        "internal_link_count": len(set(internal)),
        "external_link_count": len(set(external)),
    })
    return row


def run_crawl_lite(site_url: str) -> pd.DataFrame:
    if not site_url:
        return pd.DataFrame()
    site_url = normalize_url(site_url)
    site_netloc = urllib.parse.urlsplit(site_url).netloc
    urls = get_sitemap_urls(site_url, int(CONFIG.get("max_crawl_pages", 50)))
    print(f"Crawling {len(urls)} URLs from {site_url}")
    rows = []
    for url in tqdm(urls, desc="Crawl lite"):
        rows.append(crawl_page(url, site_netloc))
        time.sleep(float(CONFIG.get("crawl_delay_seconds", 0.5)))
    df = pd.DataFrame(rows)
    if df.empty:
        return df
    df["technical_flags"] = df.apply(lambda r: "; ".join([
        flag for flag, cond in {
            "missing_title": not bool(r.get("title")),
            "long_title": r.get("title_len", 0) > 65,
            "missing_meta_description": not bool(r.get("meta_description")),
            "missing_or_multiple_h1": r.get("h1_count", 0) != 1,
            "thin_content": r.get("word_count", 0) < 300,
            "no_schema": r.get("jsonld_count", 0) == 0,
            "images_missing_alt": r.get("images_missing_alt", 0) > 0,
            "noindex": bool(r.get("noindex")),
        }.items() if cond
    ]), axis=1)
    df.to_csv(MASTER_DIR / "technical_crawl_lite.csv", index=False)
    return df
````

### 📝 10. Automated Content Briefs
For the top-priority clusters, we generate **detailed Markdown briefs**. These include suggested outlines, target keywords, and *SEO checklists* to guide your writing team.

````python

# @title 10_CONTENT_BRIEFS_AND_SCHEMA — Deterministic content brief generation

def split_pipe(s: Any) -> List[str]:
    if pd.isna(s) or not str(s).strip():
        return []
    return [x.strip() for x in str(s).split("|") if x.strip()]


def faq_answer_stub(question: str, primary_keyword: str) -> str:
    q = question.strip().rstrip("?")
    return f"Answer this directly in 2-4 sentences. Mention {primary_keyword} naturally, define the core concept, then add a practical example or decision rule."


def build_brief_markdown(row: pd.Series, cluster_keywords: pd.DataFrame) -> str:
    primary = row["primary_keyword"]
    supporting = split_pipe(row.get("supporting_keywords", ""))
    faqs = split_pipe(row.get("faq_candidates", ""))
    content_type = row["recommended_content_type"]
    intent = row["dominant_intent"]
    title = row["recommended_title"]
    slug = row["url_slug"]
    outline = suggested_outline(primary, content_type, supporting)
    video = ""
    if CONFIG.get("include_video_angles"):
        video = f'''
## Video / Short-Form Repurpose Angle

- 60-second explainer: "The fastest way to understand {primary}"
- 3 caption hooks:
  1. "Most people get {primary} wrong because..."
  2. "Before you choose anything related to {primary}, check this."
  3. "Here is the simple version of {primary}."
'''
    md = f'''# Content Brief: {title}

## Core Target

- **Primary keyword:** {primary}
- **URL slug:** `/{slug}`
- **Dominant intent:** {intent}
- **Recommended content type:** {content_type}
- **Roadmap priority score:** {row.get("roadmap_priority_score", "")}
- **Recommended schema:** {row.get("recommended_schema", "Article, BreadcrumbList")}

## Reader Problem

The reader is searching for a clear, practical answer around **{primary}**. The page should answer the query quickly, then expand into examples, tradeoffs, and next steps.

## Suggested Title

{title}

## Suggested H1

{title}

## Supporting Keywords

{chr(10).join([f"- {kw}" for kw in supporting[:25]])}

## Suggested Outline

{outline}

## FAQ Candidates

{chr(10).join([f"- {q}?" if not q.endswith('?') else f"- {q}" for q in faqs[:10]]) if faqs else "- Add 4-6 FAQs from Search Console queries or People Also Ask research."}

## AI Search / Answer Engine Readiness Checklist

- Put a direct answer in the first 80-120 words.
- Use descriptive H2s that mirror real questions.
- Include concrete examples, not generic claims.
- Define entities, tools, brands, and acronyms when relevant.
- Add an FAQ section with concise answers.
- Add schema only when it matches visible page content.
- Include internal links to related cluster pages.

## Conversion CTA

Add a CTA that matches the intent:

- Informational: checklist, template, explainer, newsletter, calculator.
- Commercial: consultation, demo, pricing page, service inquiry.
- Comparison: decision matrix, downloadable comparison table.

{video}

## Drafting Prompt

Write a practical, non-fluffy article for the keyword `{primary}` using the outline above. Prioritize specific examples, clear definitions, concise answers, and a useful FAQ section. Avoid unsupported claims.
'''
    return md


def suggested_outline(primary: str, content_type: str, supporting: List[str]) -> str:
    base = [
        f"## What {primary} means",
        f"## Why {primary} matters",
        "## Quick answer",
        "## Key examples",
        "## Common mistakes",
        "## Step-by-step recommendation",
        "## FAQ",
        "## Next steps",
    ]
    if "comparison" in content_type:
        base = [
            "## Quick verdict",
            "## Comparison table",
            "## Option 1: strengths and tradeoffs",
            "## Option 2: strengths and tradeoffs",
            "## Which option is best for each use case",
            "## FAQ",
        ]
    elif "technical" in content_type or "tutorial" in content_type:
        base = [
            "## Quick architecture overview",
            "## Requirements and assumptions",
            "## Step-by-step implementation",
            "## Example configuration/code",
            "## Common errors and debugging",
            "## Security and maintenance notes",
            "## FAQ",
        ]
    elif "service" in content_type or "product" in content_type:
        base = [
            "## What problem this solves",
            "## Who this is for",
            "## Options and pricing factors",
            "## How to choose a provider/tool",
            "## Red flags",
            "## FAQ",
            "## CTA",
        ]
    additions = []
    for kw in supporting[:8]:
        if kw != primary and len(kw.split()) >= 3:
            additions.append(f"## Section angle: {kw}")
    return "\n".join(base + additions[:5])


def build_faq_jsonld(primary: str, faqs: List[str]) -> Dict[str, Any]:
    return {
        "@context": "https://schema.org",
        "@type": "FAQPage",
        "mainEntity": [
            {
                "@type": "Question",
                "name": q if q.endswith("?") else q + "?",
                "acceptedAnswer": {
                    "@type": "Answer",
                    "text": faq_answer_stub(q, primary)
                }
            }
            for q in faqs[:8]
        ]
    }


def generate_content_briefs(roadmap: pd.DataFrame, clustered: pd.DataFrame) -> pd.DataFrame:
    if roadmap.empty:
        return pd.DataFrame()
    manifest_rows = []
    limit = int(CONFIG.get("briefs_to_generate", 20))
    for _, row in roadmap.head(limit).iterrows():
        cid = row["cluster_id"]
        cluster_keywords = clustered[clustered["cluster_id"] == cid].copy()
        md_text = build_brief_markdown(row, cluster_keywords)
        fname = slugify(row["primary_keyword"]).replace("_", "-")
        brief_path = BRIEFS_DIR / f"{fname}.md"
        write_text(brief_path, md_text)
        faqs = split_pipe(row.get("faq_candidates", ""))
        if faqs:
            schema = build_faq_jsonld(row["primary_keyword"], faqs)
            schema_path = SCHEMA_DIR / f"{fname}.faq.jsonld"
            write_json(schema_path, schema)
        else:
            schema_path = ""
        manifest_rows.append({
            "cluster_id": cid,
            "primary_keyword": row["primary_keyword"],
            "brief_path": str(brief_path),
            "schema_path": str(schema_path),
        })
    manifest = pd.DataFrame(manifest_rows)
    manifest.to_csv(MASTER_DIR / "content_brief_manifest.csv", index=False)
    return manifest
````

This section creates **data visualizations** and a project README. It then packages all deliverables into a *single ZIP file* for easy download and client delivery.

````python

# @title 11_VISUALS_AND_EXPORT — Charts, README, manifest, zip

def make_basic_charts(deduped: pd.DataFrame, clustered: pd.DataFrame, roadmap: pd.DataFrame) -> None:
    if deduped.empty:
        return
    # Intent distribution
    intent_counts = deduped["intent"].value_counts().head(12)
    plt.figure(figsize=(10, 6))
    intent_counts.plot(kind="bar")
    plt.title("Keyword Intent Distribution")
    plt.xlabel("Intent")
    plt.ylabel("Keyword Count")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.savefig(CHARTS_DIR / "intent_distribution.png", dpi=160)
    plt.close()

    if not roadmap.empty:
        top = roadmap.head(15).sort_values("roadmap_priority_score")
        plt.figure(figsize=(10, 7))
        plt.barh(top["primary_keyword"], top["roadmap_priority_score"])
        plt.title("Top Content Roadmap Priorities")
        plt.xlabel("Priority Score")
        plt.tight_layout()
        plt.savefig(CHARTS_DIR / "content_roadmap_priorities.png", dpi=160)
        plt.close()


def make_readme(raw: pd.DataFrame, deduped: pd.DataFrame, clustered: pd.DataFrame, roadmap: pd.DataFrame, gsc: pd.DataFrame, crawl: pd.DataFrame) -> None:
    lines = [
        f"# SEOBoss Pro v2 Export — {CONFIG['project_name']}",
        "",
        f"Run ID: `{RUN_ID}`",
        f"Generated: `{datetime.now().isoformat(timespec='seconds')}`",
        "",
        "## Inputs",
        "",
        "Seed keywords:",
    ]
    lines += [f"- {kw}" for kw in CONFIG["seed_keywords"]]
    lines += [
        "",
        "## Output Summary",
        "",
        f"- Raw suggestion rows: {len(raw):,}",
        f"- Deduped keywords: {len(deduped):,}",
        f"- Keyword clusters: {clustered['cluster_id'].nunique() if not clustered.empty else 0:,}",
        f"- Roadmap items: {len(roadmap):,}",
        f"- GSC opportunity rows: {len(gsc):,}",
        f"- Crawl rows: {len(crawl):,}",
        "",
        "## Key Files",
        "",
        "- `_master/deduped_keywords.csv`",
        "- `_master/keyword_clusters.csv`",
        "- `_master/content_roadmap.csv`",
        "- `_master/money_keywords.csv`",
        "- `_master/source_summary.csv`",
        "- `content_briefs/*.md`",
        "- `schema_jsonld/*.jsonld`",
        "- `charts/*.png`",
        "",
        "## Suggested Next Action",
        "",
        "Open `_master/content_roadmap.csv`, sort by `roadmap_priority_score`, and produce the top 3 briefs first.",
    ]
    write_text(RUN_DIR / "README.md", "\n".join(lines))


def make_manifest(raw: pd.DataFrame, deduped: pd.DataFrame, clustered: pd.DataFrame, roadmap: pd.DataFrame, gsc: pd.DataFrame, crawl: pd.DataFrame) -> None:
    manifest = {
        "project_name": CONFIG["project_name"],
        "project_slug": PROJECT_SLUG,
        "run_id": RUN_ID,
        "run_dir": str(RUN_DIR),
        "seed_keywords": CONFIG["seed_keywords"],
        "counts": {
            "raw_suggestions": len(raw),
            "deduped_keywords": len(deduped),
            "clusters": int(clustered["cluster_id"].nunique()) if not clustered.empty and "cluster_id" in clustered else 0,
            "roadmap_items": len(roadmap),
            "gsc_rows": len(gsc),
            "crawl_rows": len(crawl),
        },
        "files": {
            "deduped_keywords": str(MASTER_DIR / "deduped_keywords.csv"),
            "keyword_clusters": str(MASTER_DIR / "keyword_clusters.csv"),
            "content_roadmap": str(MASTER_DIR / "content_roadmap.csv"),
            "money_keywords": str(MASTER_DIR / "money_keywords.csv"),
            "readme": str(RUN_DIR / "README.md"),
        },
        "config_snapshot": CONFIG,
    }
    write_json(RUN_DIR / "manifest.json", manifest)


def zip_run_dir() -> Optional[Path]:
    if not CONFIG.get("make_zip"):
        return None
    zip_path = PROJECT_DIR / f"{PROJECT_SLUG}_{RUN_ID}_client_pack.zip"
    if zip_path.exists():
        zip_path.unlink()
    with zipfile.ZipFile(zip_path, "w", zipfile.ZIP_DEFLATED) as zf:
        for path in RUN_DIR.rglob("*"):
            if path.is_file():
                zf.write(path, path.relative_to(RUN_DIR.parent))
    print("Zip created:", zip_path)
    return zip_path
````

### ⚡ 12. Orchestration (Run All)
This is the **main entry point** that ties everything together. Running this cell triggers the *full end-to-end workflow*, transforming your seed keywords into a complete SEO strategy.

````python

# @title 12_RUN_ALL — Productized one-shot workflow

def run_all() -> Dict[str, Any]:
    started = datetime.now()
    print("SEOBoss Pro v2 starting...")
    print("Run dir:", RUN_DIR)

    raw, deduped = run_keyword_universe()
    print(f"Keyword universe: {len(raw):,} raw rows → {len(deduped):,} deduped keywords")

    clustered, roadmap = run_clustering_and_roadmap(deduped)
    print(f"Clusters: {clustered['cluster_id'].nunique() if not clustered.empty else 0:,} | Roadmap items: {len(roadmap):,}")

    gsc = pd.DataFrame()
    if CONFIG.get("enable_gsc_csv"):
        gsc = analyze_gsc_opportunities(CONFIG.get("gsc_csv_path", ""))
        print(f"GSC opportunities: {len(gsc):,}")
    else:
        print("GSC CSV disabled/skipped.")

    crawl = pd.DataFrame()
    if CONFIG.get("enable_crawl_lite") and CONFIG.get("site_url"):
        crawl = run_crawl_lite(CONFIG.get("site_url", ""))
        print(f"Crawl rows: {len(crawl):,}")
    else:
        print("Crawl-lite disabled/skipped.")

    brief_manifest = generate_content_briefs(roadmap, clustered)
    print(f"Content briefs generated: {len(brief_manifest):,}")

    make_basic_charts(deduped, clustered, roadmap)
    make_readme(raw, deduped, clustered, roadmap, gsc, crawl)
    make_manifest(raw, deduped, clustered, roadmap, gsc, crawl)
    zip_path = zip_run_dir()

    elapsed = datetime.now() - started
    result = {
        "run_dir": str(RUN_DIR),
        "zip_path": str(zip_path) if zip_path else None,
        "deduped_keywords_csv": str(MASTER_DIR / "deduped_keywords.csv"),
        "content_roadmap_csv": str(MASTER_DIR / "content_roadmap.csv"),
        "keyword_clusters_csv": str(MASTER_DIR / "keyword_clusters.csv"),
        "elapsed_seconds": round(elapsed.total_seconds(), 2),
    }
    print("\nDONE")
    print(json.dumps(result, indent=2))
    return result

if CONFIG.get("run_now", True):
    RESULT = run_all()
else:
    print("CONFIG['run_now'] is False. Call run_all() manually when ready.")
````

````python
theme_selection = "corporate_blue.html" # @param ["default_dashboard.html", "modern_dark.html", "corporate_blue.html"]
custom_path_override = "" # @param {"type":"string"}

import base64
from pathlib import Path

# --- 0. Template & Directory Prep ---
TEMPLATES_DIR = PROJECT_DIR / "templates"
TEMPLATES_DIR.mkdir(parents=True, exist_ok=True)

DEFAULT_TEMPLATE_PATH = TEMPLATES_DIR / "default_dashboard.html"

def _ensure_default_template():
    """Writes the default HTML template to a file if it doesn't exist."""
    html_content = """<!DOCTYPE html>
<html><head><style>
    body { font-family: 'Inter', sans-serif; background: #f8fafc; color: #1e293b; padding: 40px; }
    .card { background: white; border-radius: 12px; box-shadow: 0 4px 6px -1px rgba(0,0,0,0.1); padding: 24px; margin-bottom: 24px; }
    h1 { color: #0f172a; border-bottom: 2px solid #38bdf8; padding-bottom: 10px; }
    table { width: 100%; border-collapse: collapse; margin-top: 20px; }
    th { text-align: left; background: #f1f5f9; padding: 12px; font-size: 13px; text-transform: uppercase; }
    td { padding: 12px; border-bottom: 1px solid #e2e8f0; }
    .intent-tag { background: #e0f2fe; color: #0369a1; padding: 4px 8px; border-radius: 4px; font-size: 11px; font-weight: bold; }
    .grid { display: grid; grid-template-columns: 1fr 1fr; gap: 20px; }
    img { max-width: 100%; border-radius: 8px; border: 1px solid #ddd; }
</style></head><body>
    <h1>SEO Intelligence Dashboard: {{PROJECT_NAME}}</h1>
    <p>Run ID: {{RUN_ID}} | Generated: {{DATE}}</p>
    <div class='grid'>
        <div class='card'><h3>Intent Distribution</h3><img src='{{INTENT_CHART}}'></div>
        <div class='card'><h3>Priority Roadmap</h3><img src='{{ROADMAP_CHART}}'></div>
    </div>
    <div class='card'><h3>Strategic Roadmap</h3><table><thead><tr><th>Primary Keyword</th><th>Intent</th><th>Count</th><th>Score</th></tr></thead><tbody>{{CONTENT_ROWS}}</tbody></table></div>
</body></html>"""
    if not DEFAULT_TEMPLATE_PATH.exists():
        DEFAULT_TEMPLATE_PATH.write_text(html_content, encoding='utf-8')

_ensure_default_template()

def generate_html_report(result_dict, theme_dropdown, custom_override):
    try:
        roadmap_df = pd.read_csv(result_dict['content_roadmap_csv']).head(20)
    except Exception as e:
        print(f"Error loading roadmap: {e}"); return

    def img_to_b64(path):
        p = Path(path)
        return f"data:image/png;base64,{base64.b64encode(p.read_bytes()).decode()}" if p.exists() else ""

    intent_img = img_to_b64(CHARTS_DIR / "intent_distribution.png")
    roadmap_img = img_to_b64(CHARTS_DIR / "content_roadmap_priorities.png")

    rows_html = ""
    for _, r in roadmap_df.iterrows():
        rows_html += f"<tr><td><strong>{r['primary_keyword']}</strong></td><td><span class='intent-tag'>{r['dominant_intent']}</span></td><td>{r['keyword_count']}</td><td>{r['roadmap_priority_score']}</td></tr>"

    # Selection Logic: Override > Dropdown > Default
    target_file = DEFAULT_TEMPLATE_PATH
    source_label = "Default"

    potential_paths = []
    if custom_override.strip():
        potential_paths.append((Path(custom_override), "Custom Override"))
        potential_paths.append((TEMPLATES_DIR / custom_override, "Custom Override (Templates Folder)"))
    if theme_dropdown:
        potential_paths.append((TEMPLATES_DIR / theme_dropdown, f"Dropdown: {theme_dropdown}"))

    for path_obj, label in potential_paths:
        if path_obj.exists() and path_obj.is_file():
            target_file = path_obj
            source_label = label
            break
    else:
        if custom_override.strip():
            print(f"⚠️ Warning: Custom path '{custom_override}' not found. Falling back.")

    print(f"📌 Using template: {target_file.name} (Source: {source_label})")
    raw_html = target_file.read_text(encoding='utf-8')

    final_html = raw_html.replace("{{PROJECT_NAME}}", CONFIG.get('project_name', 'SEO Project'))\
                        .replace("{{RUN_ID}}", RUN_ID)\
                        .replace("{{DATE}}", datetime.now().strftime('%Y-%m-%d'))\
                        .replace("{{INTENT_CHART}}", intent_img)\
                        .replace("{{ROADMAP_CHART}}", roadmap_img)\
                        .replace("{{CONTENT_ROWS}}", rows_html)

    out_path = RUN_DIR / "seo_intelligence_report.html"
    out_path.write_text(final_html, encoding='utf-8')

    print(f"✅ Report saved: {out_path}")
    display(HTML(final_html))

if 'RESULT' in globals():
    generate_html_report(RESULT, theme_selection, custom_path_override)
else:
    print('No results available to report.')
````

### 📊 14. Google Sheets Export
This section exports the strategic roadmap directly to a Google Sheet for collaborative tracking and project management.

````python
import gspread
from google.colab import auth
from google.auth import default
import numpy as np

def export_to_sheets(result_dict):
    try:
        # 1. Authenticate
        auth.authenticate_user()
        creds, _ = default()
        gc = gspread.authorize(creds)

        # 2. Load Data
        roadmap_df = pd.read_csv(result_dict['content_roadmap_csv'])

        # Fix: Replace NaN/Inf with empty strings or 0 to avoid JSON compliance errors
        roadmap_df = roadmap_df.replace([np.inf, -np.inf], np.nan).fillna("")

        # 3. Create Spreadsheet
        sheet_name = f"SEO Roadmap: {CONFIG.get('project_name', 'Untitled Project')} ({RUN_ID})"
        sh = gc.create(sheet_name)
        worksheet = sh.get_worksheet(0)
        worksheet.update_title("Strategic Roadmap")

        # 4. Prepare and Upload Data
        data_to_upload = [roadmap_df.columns.values.tolist()] + roadmap_df.values.tolist()
        worksheet.update('A1', data_to_upload)

        # 5. Formatting
        worksheet.format("A1:Z1", {"textFormat": {"bold": True}, "backgroundColor": {"red": 0.9, "green": 0.9, "blue": 0.9}})

        print(f"✅ Success! Google Sheet created: {sheet_name}")
        print(f"🔗 Access it here: {sh.url}")
        return sh.url

    except Exception as e:
        print(f"❌ Sheets export failed: {e}")
        return None

if 'RESULT' in globals():
    SHEET_URL = export_to_sheets(RESULT)
else:
    print('No results available to export.')
````
